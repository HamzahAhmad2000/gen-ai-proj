```python
from diffusers import StableDiffusionPipeline
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import DataLoader
from datasets import load_dataset
import torch
import torch.nn as nn

# Load Pretrained Model and Tokenizer
model_name = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_name, torch_dtype=torch.float16)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Load Dataset
dataset = load_dataset("coco_captions", split="train")
def preprocess_data(examples):
    images = [example["image"] for example in examples["images"]]
    captions = tokenizer(examples["captions"], truncation=True, padding="max_length", return_tensors="pt")
    return {"images": images, "captions": captions["input_ids"]}

dataset = dataset.map(preprocess_data)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Define Fine-Tuning Loop
optimizer = torch.optim.AdamW(pipe.unet.parameters(), lr=1e-5)
criterion = nn.CrossEntropyLoss()

for epoch in range(3):  # Number of epochs
    for batch in dataloader:
        images = batch["images"].to("cuda")
        captions = batch["captions"].to("cuda")
        optimizer.zero_grad()
        
        # Forward pass
        outputs = pipe.unet(images, captions)
        
        # Compute loss and backward pass
        loss = criterion(outputs, captions)
        loss.backward()
        optimizer.step()
        
        print(f"Epoch: {epoch}, Loss: {loss.item()}")

# Save Fine-Tuned Model
pipe.save_pretrained("./fine_tuned_model")
```
